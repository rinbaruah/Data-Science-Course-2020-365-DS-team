```python
reason_columns=pd.get_dummies(df['Reason for Absence'],drop_first=True)
reason_columns.head()
```

```python
reason_columns['check']=reason_columns.sum(axis=1) #For summing horizontally, axis =1
```

```python
reason_columns
```

```python
reason_columns['check'].sum(axis=0)
```

```python
reason_columns.drop('check',axis=1,inplace=True)
```

```python
age_dummies=pd.get_dummies(df['Age'],drop_first=True)
```

```python
age_dummies
```

```python
df.drop('Reason for Absence',axis=1,inplace=True)
```

```python
# VVI
reason_type_1=reason_columns.iloc[:,0:14].max(axis=1)
reason_type_2=reason_columns.iloc[:,14:17].max(axis=1)
reason_type_3=reason_columns.iloc[:,17:20].max(axis=1)
reason_type_4=reason_columns.iloc[:,20:].max(axis=1)
```

```python
reason_columns.iloc[:,0:14].max(axis=1)
```

```python
df=pd.concat([df,reason_type_1,reason_type_2,reason_type_3,reason_type_4],axis=1)
df
```

```python
df=df.rename(columns={0:'Reason_1',1:'Reason_2',2:'Reason_3',3:'Reason_4'})
```

```python
df.head()
```

```python
new_columns=['Reason_1', 'Reason_2', 'Reason_3',
       'Reason_4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Pets','Children',
       'Absenteeism Time in Hours' ]
```

```python
df=df[new_columns]
df
```

```python
df_reason_mod=df.copy()
#Checkpoint
```

```python
type(df.Date[0])
```

```python
df_reason_mod['Date']=pd.to_datetime(df_reason_mod['Date'],format='%d/%m/%Y')

df_reason_mod['Date']
```

```python
df['Month Value']=df_reason_mod.Date.dt.month
```

```python
df
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
#converts the date to a day of the week
```

```python
df['Day of the week']=df_reason_mod['Date'].apply(date_to_weekday)
```

```python
df.head()
```

```python
df.drop('Date',axis=1,inplace=True)
```
